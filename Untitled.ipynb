```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.wikipedia.org")

# Locate search bar using different locators
search_by_id = driver.find_element(By.ID, "searchInput")
search_by_name = driver.find_element(By.NAME, "search")
search_by_xpath = driver.find_element(By.CSS_SELECTOR, "button[type='submit']")

# Print confirmation for each locator
print(f"Search bar by ID: {search_by_id.tag_name}")
print(f"Search bar by Name: {search_by_name.tag_name}")
print(f"Search bar by CSS_SELECTOR: {search_by_xpath.tag_name}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.google.com")

# Locate search bar and input text
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys("Selenium Python")

# Click the search button
search_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.NAME, "btnK"))
)
search_button.click()

# Print the page title after search
print(driver.title)

driver.quit()
```

```python
from selenium import webdriver

driver = webdriver.Chrome()

# Navigate to two different pages
driver.get("https://www.wikipedia.org")
print(f"First page title: {driver.title}")

driver.get("https://www.python.org")
print(f"Second page title: {driver.title}")

# Navigate back
driver.back()
print(f"After back: {driver.title}")

# Navigate forward
driver.forward()
print(f"After forward: {driver.title}")

# Refresh the page
driver.refresh()
print("Page refreshed.")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize WebDriver
driver = webdriver.Chrome()

try:
    # Open the dynamic loading page
    driver.get("https://the-internet.herokuapp.com/dynamic_loading/1")

    # Start the loading process by clicking the button
    start_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.TAG_NAME, "button"))
    )
    start_button.click()

    # Wait for the element to become visible
    finish_element = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.ID, "finish"))
    )

    # Scroll to the element (optional)
    driver.execute_script("arguments[0].scrollIntoView();", finish_element)

    # Print the visible text
    print(f"Text inside the element: {finish_element.text}")

finally:
    # Close the browser
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://the-internet.herokuapp.com/javascript_alerts")

# Trigger the alert
driver.find_element(By.XPATH, "//button[text()='Click for JS Alert']").click()

# Switch to alert and print its text
alert = driver.switch_to.alert
print(f"Alert text: {alert.text}")

# Accept the alert
alert.accept()

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver = webdriver.Chrome()

try:
    # Open the Tengrinews website
    driver.get("https://tengrinews.kz")
    print("Test is done")
    # Find all news headlines and links in the main news section
    headlines = driver.find_elements(By.CSS_SELECTOR, ".main-news_top_item")
    print()
    links = driver.find_elements(By.CSS_SELECTOR, ".main-news_top_item a")

    # Print the headlines and corresponding links
    for headline, link in zip(headlines, links):
        print(f"Headline: {headline.text}")
        print(f"Link: {link.get_attribute('href')}")
        print("---")
finally:
    # Close the browser
    driver.quit()
```
